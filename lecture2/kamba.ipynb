## Part 1: Kamba Language - Sentences and Grammar

### Grammar Rules (Context-Free Grammar)

```
S   -> NP VP
NP  -> NP Adj
NP  -> N
NP  -> Det
NP  -> N Det
VP  -> VP Adj
VP  -> V
VP  -> V NP
Adj -> Prep NP
Adj -> NP
```

````python
# 1. Akakoma
# 2. Osa kalamu
# 3. Kiandiki Kya Yoana kyai mesani
# 4. Andu aume aa athuku makatema miti mingi 
# 5. Anainya menyu on the matineekoma
````

## Part 2: NLTK Context-Free Grammar (CFG) & Sentence Generation



````python
import nltk
from nltk import CFG
from nltk.parse.generate import generate
from nltk.parse import RecursiveDescentParser

kamba_grammar = CFG.fromstring("""
    S -> NP VP
    NP -> NP Adj
    NP -> N
    NP -> Det
    NP -> N Det
    VP -> VP Adj
    VP -> V
    VP -> V NP
    Adj -> Prep NP
    Adj -> NP
    
    Det -> 'yila' | 'isu' | 'inya'
    N -> 'mwanake' | 'kaveti' | 'mutumia' | 'mwiitu' | 'mwana' | 'nyama' | 'ivuku' | 'maii' | 'mboo' | 'iio' | 'nzeve' | 'kiveti'
    V -> 'nikusya' | 'niwaasa' | 'asyaa' | 'anasoma' | 'aithukia' | 'nikwosa' | 'akoma' | 'aenda' | 'ninasya' | 'niwaiyuia'
    Prep -> 'na' | 'wa' | 'kwa'
""")

print("Kamba CFG Grammar:")
print(kamba_grammar)
print(f"\nNumber of productions: {len(kamba_grammar.productions())}")
````

````python
# Generate sentences from the grammar (limit depth to avoid infinite recursion)
print("=" * 60)
print("SENTENCES GENERATED FROM THE KAMBA CFG")
print("=" * 60)

generated = list(generate(kamba_grammar, depth=4, n=20))
for i, sent in enumerate(generated, 1):
    print(f"{i}. {' '.join(sent)}")

print(f"\nTotal sentences generated (depth=4, max 20): {len(generated)}")
````

````python
# Parse some Kamba sentences using the grammar
parser = RecursiveDescentParser(kamba_grammar)

test_sentences = [
    "mwanake nikusya nyama",
    "mwiitu anasoma ivuku",
    "kaveti niwaasa maii",
    "mutumia asyaa iio",
    "mwana akoma"
]

print("=" * 60)
print("PARSING KAMBA SENTENCES")
print("=" * 60)

for sent in test_sentences:
    tokens = sent.split()
    print(f"\nSentence: {sent}")
    try:
        trees = list(parser.parse(tokens))
        if trees:
            for tree in trees:
                print(tree)
        else:
            print("  -> No parse found (sentence not in grammar)")
    except ValueError as e:
        print(f"  -> Parse error: {e}")
````

## Part 3: Dependency Grammar for Kamba Sentences

````python
from nltk.parse import DependencyGraph

dep1 = "mwanake\tN\t2\tnsubj\nnikusya\tV\t0\tROOT\nnyama\tN\t2\tdobj\n"
dep2 = "kaveti\tN\t2\tnsubj\nniwaasa\tV\t0\tROOT\nmwii\tN\t2\tdobj\n"
dep3 = "mutumia\tN\t2\tnsubj\nasyaa\tV\t0\tROOT\niio\tN\t2\tdobj\n"
dep4 = "mwiitu\tN\t2\tnsubj\nanasoma\tV\t0\tROOT\nivuku\tN\t2\tdobj\n"
dep5 = "kaveti\tN\t2\tnsubj\naithukia\tV\t0\tROOT\nmwana\tN\t2\tdobj\n"

dep_sentences = {
    "Mwanake nikusya nyama": dep1,
    "Kaveti niwaasa mwii": dep2,
    "Mutumia asyaa iio": dep3,
    "Mwiitu anasoma ivuku": dep4,
    "Kaveti aithukia mwana": dep5
}

for sent_text, dep_str in dep_sentences.items():
    print(f"\nSentence: {sent_text}")
    print("-" * 40)
    dg = DependencyGraph(dep_str)
    print(dg.to_conll(4))
    dg_tree = dg.tree()
    print("Tree:", dg_tree)
    dg_tree.pretty_print()
````

## Part 4: Analyzing and Processing Text with spaCy (English)

````python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """Barack Obama was the 44th president of the United States. 
He was born in Honolulu, Hawaii on August 4, 1961. 
Obama graduated from Columbia University and Harvard Law School. 
He served as a U.S. senator from Illinois before becoming president in January 2009.
Apple Inc. is planning to open a new office in Nairobi, Kenya next year."""

doc = nlp(text)
print(f"Tokens: {len(doc)}, Sentences: {len(list(doc.sents))}")
````

````python
# 4.1 Tokenization
print("=" * 60)
print("TOKENIZATION")
print("=" * 60)
print(f"{'Token':<20} {'Is Alpha':<10} {'Is Stop':<10} {'Is Punct':<10}")
print("-" * 50)
for token in doc[:25]:  # First 25 tokens
    print(f"{token.text:<20} {str(token.is_alpha):<10} {str(token.is_stop):<10} {str(token.is_punct):<10}")
````

````python
# 4.2 Sentence Segmentation
print("=" * 60)
print("SENTENCE SEGMENTATION")
print("=" * 60)
for i, sent in enumerate(doc.sents, 1):
    print(f"\nSentence {i}: {sent.text.strip()}")
````

````python
# 4.3 Part-of-Speech (POS) Tagging
print("=" * 60)
print("POS TAGGING")
print("=" * 60)
print(f"{'Token':<20} {'POS':<10} {'Fine POS':<15} {'Explanation'}")
print("-" * 70)
for token in doc[:30]:
    print(f"{token.text:<20} {token.pos_:<10} {token.tag_:<15} {spacy.explain(token.tag_)}")
````

````python
# 4.4 Lemmatization
print("=" * 60)
print("LEMMATIZATION")
print("=" * 60)
print(f"{'Token':<20} {'Lemma':<20} {'POS':<10}")
print("-" * 50)
for token in doc:
    if token.text != token.lemma_ and token.is_alpha:
        print(f"{token.text:<20} {token.lemma_:<20} {token.pos_:<10}")
````

````python
# 4.5 Dependency Parsing
print("=" * 60)
print("DEPENDENCY PARSING")
print("=" * 60)
print(f"{'Token':<20} {'Dep':<15} {'Head':<20} {'Children'}")
print("-" * 70)
first_sent = list(doc.sents)[0]
for token in first_sent:
    children = [child.text for child in token.children]
    print(f"{token.text:<20} {token.dep_:<15} {token.head.text:<20} {children}")
````

````python
# 4.6 Named Entity Recognition (NER)
print("=" * 60)
print("NAMED ENTITY RECOGNITION (NER)")
print("=" * 60)
print(f"{'Entity':<30} {'Label':<15} {'Description'}")
print("-" * 70)
for ent in doc.ents:
    print(f"{ent.text:<30} {ent.label_:<15} {spacy.explain(ent.label_)}")
````

````python
# 4.7 Noun Chunks
print("=" * 60)
print("NOUN CHUNKS")
print("=" * 60)
print(f"{'Chunk':<30} {'Root':<15} {'Root Dep':<15} {'Head'}")
print("-" * 70)
for chunk in doc.noun_chunks:
    print(f"{chunk.text:<30} {chunk.root.text:<15} {chunk.root.dep_:<15} {chunk.root.head.text}")
````

````python
# 4.8 Word Vectors and Similarity
print("=" * 60)
print("WORD SIMILARITY")
print("=" * 60)

tokens_to_compare = nlp("president senator university school")
for token1 in tokens_to_compare:
    for token2 in tokens_to_compare:
        if token1.text != token2.text:
            sim = token1.similarity(token2)
            print(f"{token1.text:<15} <-> {token2.text:<15} similarity: {sim:.4f}")
````

## Part 5: Web Scraping with Requests and BeautifulSoup

````python
import requests
from bs4 import BeautifulSoup

url = 'https://www.python.org/'
page = requests.get(url)

print(f"URL: {url}")
print(f"Status Code: {page.status_code}")
print(f"Content Type: {page.headers['Content-Type']}")
print(f"Page size: {len(page.content)} bytes")
````

````python
soup = BeautifulSoup(page.text, 'html.parser')

print(f"Page Title: {soup.title.text}")
print()

links = soup.find_all('a')
print(f"Total number of links: {len(links)}")
print("\nFirst 10 links:")
for i, link in enumerate(links[:10], 1):
    href = link.get('href', 'N/A')
    text = link.get_text(strip=True)[:50]
    print(f"  {i}. Text: '{text}' -> {href}")
````

````python
for tag in ['h1', 'h2', 'h3']:
    headings = soup.find_all(tag)
    if headings:
        print(f"\n{tag.upper()} headings ({len(headings)}):")
        for h in headings[:5]:
            print(f"  - {h.get_text(strip=True)[:80]}")
````

````python
paragraphs = soup.find_all('p')
print(f"Total paragraphs: {len(paragraphs)}")
print("\nFirst 5 paragraphs:")
for i, p in enumerate(paragraphs[:5], 1):
    text = p.get_text(strip=True)
    if text:
        print(f"\n{i}. {text[:200]}")
````

````python
images = soup.find_all('img')
print(f"Total images: {len(images)}")
for i, img in enumerate(images[:10], 1):
    src = img.get('src', 'N/A')
    alt = img.get('alt', 'N/A')
    print(f"  {i}. Alt: '{alt}' -> {src}")
````

## Part 6: Word Sense Disambiguation (WSD) and Coreference Resolution

### 6.1 Word Sense Disambiguation (WSD)

````python
from nltk.wsd import lesk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize

sentences_bank = [
    "I went to the bank to deposit my money",
    "The river bank was covered with beautiful flowers",
    "We sat on the bank of the lake watching the sunset"
]

print("--- WSD for the word 'bank' ---")
for sent in sentences_bank:
    tokens = word_tokenize(sent)
    sense = lesk(tokens, 'bank')
    print(f"\nSentence: {sent}")
    if sense:
        print(f"  Sense: {sense.name()}")
        print(f"  Definition: {sense.definition()}")
    else:
        print("  No sense found")

sentences_bat = [
    "He hit the ball with a bat",
    "The bat flew out of the cave at night"
]

print("\n--- WSD for the word 'bat' ---")
for sent in sentences_bat:
    tokens = word_tokenize(sent)
    sense = lesk(tokens, 'bat')
    print(f"\nSentence: {sent}")
    if sense:
        print(f"  Sense: {sense.name()}")
        print(f"  Definition: {sense.definition()}")
    else:
        print("  No sense found")
````

````python
for word in ['bank', 'bat', 'plant', 'bass']:
    synsets = wn.synsets(word)
    print(f"\nWord: '{word}' has {len(synsets)} senses:")
    for i, ss in enumerate(synsets[:5], 1):
        print(f"  {i}. {ss.name()}: {ss.definition()}")
    if len(synsets) > 5:
        print(f"  ... and {len(synsets)-5} more senses")
````

````python
plant_sentences = [
    "The plant in the garden is growing well",
    "They built a new manufacturing plant in the city"
]

print("--- WSD for the word 'plant' ---")
for sent in plant_sentences:
    tokens = word_tokenize(sent)
    sense = lesk(tokens, 'plant')
    print(f"\nSentence: {sent}")
    if sense:
        print(f"  Sense: {sense.name()}")
        print(f"  Definition: {sense.definition()}")

bass_sentences = [
    "He plays the bass guitar in the band",
    "We caught a large bass in the lake"
]

print("\n--- WSD for the word 'bass' ---")
for sent in bass_sentences:
    tokens = word_tokenize(sent)
    sense = lesk(tokens, 'bass')
    print(f"\nSentence: {sent}")
    if sense:
        print(f"  Sense: {sense.name()}")
        print(f"  Definition: {sense.definition()}")
````

### 6.2 Coreference Resolution

````python
coref_text = """Barack Obama was born in Hawaii. He graduated from Harvard Law School.
The former president served two terms. Obama left office in 2017."""

doc_coref = nlp(coref_text)

print("Text:")
print(coref_text)

print("\n--- Named Entities Found ---")
for ent in doc_coref.ents:
    print(f"  {ent.text:<25} -> {ent.label_:<10} ({spacy.explain(ent.label_)})")

print("\n--- Coreference Chains ---")
print("Chain 1: 'Barack Obama' = 'He' = 'The former president' = 'Obama'")
print("Chain 2: 'Hawaii'")
print("Chain 3: 'Harvard Law School'")

print("\n--- Pronoun Resolution ---")
for token in doc_coref:
    if token.pos_ == 'PRON':
        print(f"  Pronoun: '{token.text}' at position {token.i}")
        print(f"    Head: '{token.head.text}', Dep: {token.dep_}")
        antecedent = None
        for prev_token in doc_coref[:token.i]:
            if prev_token.pos_ == 'PROPN':
                antecedent = prev_token.text
        if antecedent:
            print(f"    Likely antecedent: '{antecedent}'")
````
