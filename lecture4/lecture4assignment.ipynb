```python
# Core imports for EDA and preprocessing
import pandas as pd
import numpy as np
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Twitter training CSV — initial load & EDA
This notebook loads `lecture4/twitter_training.csv`, inspects schema, checks label distribution, duplicates, missing values, and plots basic distributions to guide cleaning decisions.

```python
# Load CSV with fallback encoding and set column names (no header in file)
path = 'lecture4/twitter_training.csv'
try:
    df = pd.read_csv(path, header=None, names=['id','topic','label','text'], encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv(path, header=None, names=['id','topic','label','text'], encoding='latin-1')
df.shape
df.head(8)
```

```python
# Basic info and label counts
print('Info:')
display(df.info())
print('
Missing text values:' , df['text'].isnull().sum())
print('
Label distribution:')
print(df['label'].value_counts(dropna=False))
# Add a text length column for EDA
df['text_length'] = df['text'].astype(str).str.len()
df['text_length'].describe()
```

```python
# Duplicates and empty/short texts
print('Exact duplicate text rows:', df.duplicated(subset=['text']).sum())
print('Rows with empty or very short text (<3 chars):', (df['text'].astype(str).str.strip().str.len() < 3).sum())
# Show a few duplicate examples
dups = df[df.duplicated(subset=['text'], keep=False)].sort_values('text')
dups.head(6)
```

```python
# Quick visualizations: label counts and text length distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, y='label', order=df['label'].value_counts().index)
plt.title('Label distribution')
plt.tight_layout()
plt.show()
plt.figure(figsize=(8,4))
sns.histplot(df['text_length'].dropna(), bins=40)
plt.title('Text length distribution')
plt.tight_layout()
plt.show()
```

```python
# Lightweight cleaning preview: drop exact duplicate texts and very short/empty texts
clean = df.drop_duplicates(subset=['text']).copy()
clean = clean[clean['text'].astype(str).str.strip().str.len() >= 3].reset_index(drop=True)
print('Before:', df.shape, 'After (preview):', clean.shape)
# Save a small cleaned preview for review
clean.sample(n=100, random_state=42).to_csv('lecture4/twitter_training_clean_preview.csv', index=False)
print('Wrote lecture4/twitter_training_clean_preview.csv')
```
